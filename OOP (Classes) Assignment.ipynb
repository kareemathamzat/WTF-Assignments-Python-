```python
class Player:
    def __init__(self,num):
        self.__num = num
        self.__score = 0
        self.__answer = -1
        
    def set_answer(self):
        ans = int(input(f'Player {self.__num}, Enter your first guess[1,6]; '))
        self.__answer = ans
        
    def get_answer(self):
        if self.__answer == -1:
            print(f'Player #{self.__num}, Enter your first guess:')
        else:
            return self.__answer
        
    def set_score(self):
        self.__score += 1
        
    def get_score(self):
        return self.__score
        
        
```

```python
from random import randint
player1 = Player(1)
player2 = Player(2)

for i in range(1,7):
    print(f'round #{i}')
    guessed_num = randint(1,6)
    
    player1.set_answer()
    player2.set_answer()
    
    if guessed_num == player1.get_answer():
        player1.set_score()
    if guessed_num == player2.get_answer():
        player2.set_score()
        
    print(f'The right number is {guessed_num}')
    print(f'Player 1 scored {player1.get_score()} and Player 2 scored {player2.get_score()}')
    print('\n')
        
```
