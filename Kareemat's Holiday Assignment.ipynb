Write a program that plays a “hi-low” number guessing game. The program starts by generating a random number hidden from the user that is between 0 and 100. The user then attempts to guess the number, getting hints as to which direction (bigger or smaller, higher or lower) to go on the next guess. The game can end in one of two ways:
<br>-The user can correctly guess the number.
<br>-The user can quit playing by entering a number out of the range of 0–100

```python
from random import randint
x = randint(0,100)
y = None

while x != y:
    y = int(input('Enter your guess -> '))
    if y < 0 or y > 100:
         print('You guessed a number out of the range of 0-100')
         break
    elif y < x:
         print('The number you guessed is smaller than the generated number')
    elif y > x:
         print('The number you guessed is bigger than the generated number')
else:
    print('The number you guessed is equal to the generated number')
    print('\n')
```

A perfect number is an integer whose sum of integer divisors (excluding the number itself) add up to the number. Following
are the first four perfect numbers:
<br> 6 = 1 + 2 + 3
<br> 28 = 1 + 2 + 4 + 7 + 14
<br> 496 = 1 + 2 + 4 + 8 + 16 + 31 + 62 + 124 + 248
<br> 8128 = 1 + 2 + 4 + 8 + 16 + 32 + 64 + 127 + 254 + 508 + 1016 + 2032 + 4064

Numbers whose sum of divisors is larger than the number are called abundant numbers. Those whose sum is less than the number
are called deficient numbers. Perfect numbers have other interesting properties, but for now we are interested only in
writing a program that sorts integers into one of three categories: perfect, abundant, or deficient. Write a program 
so that we get the following results.

```python
def perfect_number(c: int):
    divisors_sum = sum(i for i in range(1, c) if c % i == 0)
    if divisors_sum == c:
        return(str(c)+ " is a perfect number")
    else:
        print(str(c)+ " is not a perfect number")
        if divisors_sum > c:
            return(str(c)+ " is an abundant number")
        else:
            return(str(c)+ " is a deficient number")
 

c = int(input("Enter a number ->"))
perfect_number(c)

```

Continuing on task 1, we would like to check a whole range of number and decide what each number is: abundant, deficient, or perfect. In this way we can look for more perfect numbers. Write a program so that we get the following results:


```python
n = int(input('Enter a number for the upper range; '))
for i in range(2,n+1):
    
    divisors_sum = sum([j for j in range(1,i) if i%j == 0])
    if divisors_sum == i:
        print(str(i)+ " is a perfect number")
    elif divisors_sum > i:
        print(str(i)+ " is an abundant number")
    else:
        print(str(i)+ " is a deficient number")
```
